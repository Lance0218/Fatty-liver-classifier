># train dataset
>>## separate in four period by steatosis:
>>>### Normal : < 5 %
>>>### Mild : 5 ~ 33 %
>>>### Moderate : 33 ~ 66 %
>>>### Severe : > 66 %

```python
# Box plot
# import models
import xlrd
import numpy as np
import matplotlib.pyplot as plt

# load data
## Pathology for train
data = xlrd.open_workbook("Pathology.xlsx").sheets()[0]

y_train = np.zeros((111, 3))
y_train[:, 0] = data.col_values(3)  # y0 = Integrated backscatter
y_train[:, 1] = data.col_values(4)  # y1 = Q factor (HHT)
y_train[:, 2] = data.col_values(6)  # y2 = Homogeneity factor

### group
y0 = []
y1 = []
y2 = []
y3 = []
nrows = data.nrows
for i in range(nrows):  # X = 'T' or 'F' from Pathology data 
    if data.cell(i,1).value < 5:
        y0.append(y_train[i, :].tolist())
    elif data.cell(i,1).value <= 33:
        y1.append(y_train[i, :].tolist())
    elif data.cell(i,1).value <= 66:
        y2.append(y_train[i, :].tolist())
    else:
        y3.append(y_train[i, :].tolist())

### list to array
y0 = np.asarray(y0)
y1 = np.asarray(y1)
y2 = np.asarray(y2)
y3 = np.asarray(y3)

# def box_plot
def box_plot(i):
    color_fc = ['pink', 'lightyellow', 'lightgreen', 'lightblue']
    color_eg = ['red', 'yellow', 'green', 'blue']
    plt.figure(figsize = (15, 10))
    bplot = plt.boxplot(x = [y0[:,i],  y1[:, i], y2[:, i], y3[:, i]], 
                patch_artist = True,
                medianprops = {'color':'black'})
    for patch, color_fc, color_eg in zip(bplot['boxes'], color_fc, color_eg):
        patch.set_facecolor(color_fc)
        patch.set_edgecolor(color_eg)
    plt.title('{0} training boxplot'.format(categories[i]), size=30)
    plt.xlabel('Disease stage', size = 25)
    plt.ylabel('{0}'.format(features[i]), size = 25)
    plt.xticks(x, ['Normal', 'Mild', 'Moderate', 'Severe'], size = 15)
    plt.yticks(size = 15)
    plt.show()
    print(bplot['boxes'])
    
# plot
x = np.arange(1,5)
categories = ['Integrated backscatter', 'HHT Q factor', 'Homogeneity factor']
features = ['Integrated backscatter (dB)', 'HHT Q factor', 'Homogeneity factor']
for i in range(3):
    box_plot(i)
```
