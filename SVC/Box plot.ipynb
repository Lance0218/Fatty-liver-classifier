>>>### Normal : < 5 %
>>>### Mild : 5 ~ 33 %
>>>### Moderate : 33 ~ 66 %
>>>### Severe : > 66 %

```python
# Box plot
# import models
import xlrd
import numpy as np
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot

# open offline plotly
init_notebook_mode(connected=True)

# load data
## Pathology for train
data = xlrd.open_workbook("Pathology.xlsx").sheets()[0]

y_train = np.zeros((111, 3))
y_train[:, 0] = data.col_values(3)  # y0 = Integrated backscatter
y_train[:, 1] = data.col_values(4)  # y1 = Q factor (HHT)
y_train[:, 2] = data.col_values(6)  # y2 = Homogeneity factor

### group
y0 = []
y1 = []
y2 = []
y3 = []
nrows = data.nrows
for i in range(nrows):  # X = 'T' or 'F' from Pathology data 
    if data.cell(i,1).value < 5:
        y0.append(y_train[i, :].tolist())
    elif data.cell(i,1).value <= 33:
        y1.append(y_train[i, :].tolist())
    elif data.cell(i,1).value <= 66:
        y2.append(y_train[i, :].tolist())
    else:
        y3.append(y_train[i, :].tolist())

### list to array
y0 = np.asarray(y0)
y1 = np.asarray(y1)
y2 = np.asarray(y2)
y3 = np.asarray(y3)

# def box_plot
def box_plot(i):
    trace0 = go.Box(y = y0[:, i], name = 'Normal', marker = dict(color = 'red'))
    trace1 = go.Box(y = y1[:, i], name = 'Mild', marker = dict(color = 'yellow'))
    trace2 = go.Box(y = y2[:, i], name = 'Moderate', marker = dict(color = 'green'))
    trace3 = go.Box(y = y3[:, i], name = 'Severe', marker = dict(color = 'blue'))
    layout=go.Layout(title = '{0}'.format(categories[i]),
                     titlefont = {'size':'30'},
                     xaxis = {'title':'Disease stage', 'titlefont':{'size':'20'}},
                     yaxis = {'title':'Parameter value', 'titlefont':{'size':'20'}})
    traces = [trace0, trace1, trace2, trace3]
    fig = go.Figure(data=traces, layout=layout)
    iplot(fig)

# plot
categories = ['Integrated backscatter', 'HHT Q factor', 'Homogeneity factor']
for i in range(3):
    box_plot(i)
```
