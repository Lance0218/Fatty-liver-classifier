```python
import pandas as pd
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot

# open offline plotly
init_notebook_mode(connected=True)

# create dataframes
df = pd.DataFrame({'features' : ['IB', 'HHT', 'HF', 'IB + HHT', 'IB + HF', 'HHT + HF', 'IB + HHT + HF'], 
                   'Sensitivity (%)' : [92.86, 67.86, 94.64, 83.93, 92.86, 87.50, 92.86],
                   'Specificity (%)' : [44.44, 72.22, 38.89, 77.78, 55.56, 50.00, 66.67],
                   'LR+' : [1.67, 2.44, 1.55, 3.78, 2.09, 1.75, 2.79],
                   'LR-' : [0.16, 0.45, 0.14, 0.21, 0.13, 0.25, 0.11],
                   'PPV (%)' : [83.87, 88.37, 82.81, 92.16, 86.67, 84.48, 89.66],
                   'NPV (%)' : [66.67, 41.94, 70.00, 60.87, 71.43, 56.25, 75.00],
                   'Accuracy (%)' : [81.08, 68.92, 81.08, 82.43, 83.78, 78.38, 86.49],
                   'AUROC' : [0.8651, 0.8006, 0.7927, 0.8839, 0.8819, 0.8442, 0.8929],})

# def Bar plot
def bar_plot(i):
    layout=go.Layout(title = '{0} Compare of Feature Select'.format(Parameters[i]),
                     titlefont = {'size':'30'},
                     xaxis = {'title':'Feature Select', 'titlefont':{'size':'20'}},
                     yaxis = {'title':'{0}'.format(Parameters[i]), 'titlefont':{'size':'20'}})
    data = [go.Bar(x = df['features'], y = df[Parameters[i]])]
    fig = go.Figure(data = data, layout = layout)
    iplot(fig)

#plot
Parameters = ['Sensitivity (%)', 'Specificity (%)', 'LR+', 'LR-', 'PPV (%)', 'NPV (%)', 'Accuracy (%)', 'AUROC']
for i in range(8):
    bar_plot(i)
```
