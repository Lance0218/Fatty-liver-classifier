# Pathology data information

## show data

```python
import pandas as pd
import numpy as np
import xlrd
df = pd.read_excel('raw_score20170306 recheck.xlsx', header = 0)
df.head()
```

```python
df.columns.values
```

```python
df.loc[:, ['matnumber', 'DateARFI', 'Birth', 'Sex', 'fatty']]
```

## data cleaning

```python
# values standardization
from dateutil import relativedelta
for i in df.index:
    df.loc[i, 'Age'] = relativedelta.relativedelta(df['DateARFI'][i], df['Birth'][i]).years
df['Sex'].replace({'F':0, 'M':1}, inplace = True)
df['fatty'].replace({'1(2011)':1, '<5':5, '10 (<5%)':5, '20 (<5%)':5, 'no fatty':0}, inplace = True)
df.loc[:, ['matnumber', 'Age', 'Sex', 'fatty']]
```

## group in four stages

```python
df0 = pd.DataFrame(columns = ['Age', 'Sex', 'fatty'])
df1 = pd.DataFrame(columns = ['Age', 'Sex', 'fatty'])
df2 = pd.DataFrame(columns = ['Age', 'Sex', 'fatty'])
df3 = pd.DataFrame(columns = ['Age', 'Sex', 'fatty'])
for i in range(len(df.index)):
    if int(df['fatty'][i]) < 5:
        df0 = df0.append({'Age': df['Age'][i], 'Sex': df['Sex'][i], 'fatty': df['fatty'][i]}, ignore_index=True)
    elif int(df['fatty'][i]) <= 33:
        df1 = df1.append({'Age': df['Age'][i], 'Sex': df['Sex'][i], 'fatty': df['fatty'][i]}, ignore_index=True)
    elif int(df['fatty'][i]) < 66:
        df2 = df2.append({'Age': df['Age'][i], 'Sex': df['Sex'][i], 'fatty': df['fatty'][i]}, ignore_index=True)
    elif int(df['fatty'][i]) <= 100:
        df3 = df3.append({'Age': df['Age'][i], 'Sex': df['Sex'][i], 'fatty': df['fatty'][i]}, ignore_index=True)
```

## calculate patient information

```python
dfx= [df0, df1, df2, df3]
for i in range(4):
    print('df{0}'.format(i))
    print('Sample size : {0}'.format(len(dfx[i])))
    print('Age : {0:.2f} ± {1:.2f} ({2:.0f} - {3:.0f})'.format(dfx[i]['Age'].mean(), dfx[i]['Age'].std(), dfx[i]['Age'].min(), dfx[i]['Age'].max()))
    print('Sex : {0} woman & {1} man'.format(len(dfx[i]['Sex']) - dfx[i]['Sex'].sum(), dfx[i]['Sex'].sum()))
```
